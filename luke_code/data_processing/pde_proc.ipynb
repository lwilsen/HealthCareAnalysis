```python
pde = pd.read_csv('../data/pde_raw.csv', sep = '|')
```

```python
pde['YR'] = pd.to_datetime(pde['SRVC_DT']).dt.year
```

```python
pde["BENE_ID"].unique()
pde[pde["BENE_ID"] == -10000010254618]

cost_per_year = pde.groupby(["BENE_ID", "YR"])['TOT_RX_CST_AMT'].sum().reset_index(name = "TOT_RX_CST_AMT")

#Wide table below
cost_per_year.pivot_table(index = "BENE_ID", columns="YR", values="TOT_RX_CST_AMT").fillna(0)

cost_per_year.to_csv("../data/rx_cost_per_year.csv", index = False)
```
